## Overview of Seaborn Plotting Functions

```python

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from IPython.display import display, Markdown, Latex

import seaborn as sns

%matplotlib inline
```

### Similar Functions for Similar Tasks

```python
penguins = sns.load_dataset("penguins")  #seaborn has built-in datasets
display(Markdown('###Penguin Data'))
display(penguins.head())
print(penguins.shape)
```

```python
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple= 'stack')
```

```python
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

above is figure level plots, --matplotlib.pyplot.Axes object

below is figure level -- displot, relplot, catplot  -- Factet Grid object

If you want to do stufff to facet grid objects, you have to dig into the matplotlib stuff, find the axes of the figure...

```python
plot = sns.displot(data = penguins, x = 'body_mass_g', hue = "sex", multiple = 'dodge')
```

```python
#sns.histplot()
#figure level plots call plotting functions, so you might have to go back tot the original plotting function to identify variables
```

```python
sns.displot(data = penguins, x = 'body_mass_g', hue = "sex", multiple = 'stack', kind = 'kde')
```

```python
working = sns.catplot(data = penguins, x = 'species', y = 'body_mass_g', kind = 'violin')
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="species")
```
