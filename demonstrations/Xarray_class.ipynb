```python
count_df = pd.read_csv(RAW_COUNT_PATH, sep = "\t", index_col = 0)
display(Markdown('## Count Data'))
print(count_df.shape)
display(count_df.head())
```

```python
count_array = xr.Dataset(
    {"counts": (("Gene", "Sample"), count_df.values)},
    coords= {
        "Sample": count_df.columns.values,
        "Gene": count_df.index.values
    }
)
count_array = count_array.transpose()
count_array
```

```python
label_df = pd.read_csv(LABEL_PATH, sep = '\t', index_col = 1)
label_df['genotype'].str.split(" ", expand = True).iloc[:,0]
#label_df['time'] = label_df['time'].str.split(" ", expand = True).iloc[:, 0].astype
label_df.index.rename('Sample', inplace = True)
label_df.head()
```

```python
label_df.index.intersection(count_df.columns.values)
```

```python
label_ds = label_df.to_xarray()
label_ds
```

```python
ds = label_ds.merge(count_array)
ds
```

```python
ds.to_netcdf('oryza_sativa_gem.nc')
```
